```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
df=school_data_complete.loc[:, "school_name"]
total_schools=len(df.unique())

total_students=len(school_data_complete)

budget=school_data_complete.loc[:, "budget"].unique().sum()

average_math_score=school_data_complete.loc[:, "math_score"].mean()
average_reading_score=school_data_complete.loc[:, "reading_score"].mean()

school_data_scores=school_data_complete.loc[:, ["math_score", "reading_score"]]

m=0
n=0

for index, row in school_data_scores.iterrows():
    if row["math_score"]  >= 70:
        n+=1
    if row["reading_score"] >= 70:
        m+=1 

Passing_Math = 100*n/total_students
Passing_Reading = 100*m/total_students

Overall_Passing_Rate = (average_math_score+average_reading_score)/2

d={'Total Schools': [total_schools], 
   'Total Students': [total_students], 
   'Total Budget':[f"${budget}"], 
   'Average Math Score': [average_math_score],
   'Average Reading Score': [average_reading_score], 
   '% Passing Math': [Passing_Math],
   '% Passing Reading': [Passing_Reading],
   '% Overall Passing Rate':[Overall_Passing_Rate]
  }
df=pd.DataFrame(data=d)
df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
schools=school_data_complete.loc[:, "school_name"].unique()
Index=[]
school_type =[]
total_students =[]
budget = []
average_math_score = []
average_reading_score = []
passing_math = []
passing_reading = []
overall_passing_rate = []

for school in schools:
    school_data=school_data_complete.loc[school_data_complete['school_name']==school]
    school_data_scores=school_data.loc[:, ["math_score", "reading_score"]]
    
    m=0
    n=0
    for index, row in school_data_scores.iterrows():
        if row["math_score"]  >= 70:
            n+=1
        if row["reading_score"] >= 70:
            m+=1
    passing_math.append(100*n/len(school_data))
    passing_reading.append(100*m/len(school_data))
    overall_passing_rate.append(50*(n+m)/len(school_data))
    
    Index.append(school)
    school_type.append(school_data.loc[:, 'type'].unique()[0])
    total_students.append(len(school_data))
    budget.append(school_data.loc[:, "budget"].unique()[0])
    average_math_score.append(school_data.loc[:, "math_score"].mean())
    average_reading_score.append(school_data.loc[:, "reading_score"].mean())
    
 
    
per_student_budget = [i / j for i, j in zip(budget, total_students)] 

df=pd.DataFrame(
    {"": Index,
     "School Type": school_type,
     "Total Students": total_students,
     "Total School Budget": budget,
     "Per Student Budget": per_student_budget,
     "Average Math Score": average_math_score,
     "Average Reading Score": average_reading_score,
     "% Passing Math": passing_math, 
     "% Passing Reading": passing_reading,
     "% Overall Passing Rate": overall_passing_rate 
    })


df_top_schools = df.sort_values(by=['% Overall Passing Rate'], ascending=False).set_index('').head().style.format({"Total School Budget": "${:.2f}", "Per Student Budget": "${:.2f}"})

df_top_schools
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
schools=school_data_complete.loc[:, "school_name"].unique()
grades=['9th', '10th', '11th', '12th']

math_scores = []
for school in schools:
    school_data=school_data_complete.loc[school_data_complete['school_name']==school]
    math_score = []
    for grade in grades:
        school_grade_data=school_data.loc[school_data_complete['grade']==grade]
        math_score.append(school_grade_data.loc[:, "math_score"].mean())
    math_scores.append(math_score)   
        
df=pd.DataFrame(data=math_scores).set_index(schools).sort_index(ascending=True)
df.columns = grades
df
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
schools=school_data_complete.loc[:, "school_name"].unique()
grades=['9th', '10th', '11th', '12th']

reading_scores = []
for school in schools:
    school_data=school_data_complete.loc[school_data_complete['school_name']==school]
    reading_score = []
    for grade in grades:
        school_grade_data=school_data.loc[school_data_complete['grade']==grade]
        reading_score.append(school_grade_data.loc[:, "reading_score"].mean())
    reading_scores.append(reading_score)   
        
df=pd.DataFrame(data=reading_scores).set_index(schools).sort_index(ascending=True)
df.columns = grades
df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
schools=school_data_complete.loc[:, "school_name"].unique()
Index=[]
school_type =[]
total_students =[]
budget = []
average_math_score = []
average_reading_score = []
passing_math = []
passing_reading = []
overall_passing_rate = []

for school in schools:
    school_data=school_data_complete.loc[school_data_complete['school_name']==school]
    school_data_scores=school_data.loc[:, ["math_score", "reading_score"]]
    
    m=0
    n=0
    for index, row in school_data_scores.iterrows():
        if row["math_score"]  >= 70:
            n+=1
        if row["reading_score"] >= 70:
            m+=1
    passing_math.append(100*n/len(school_data))
    passing_reading.append(100*m/len(school_data))
    overall_passing_rate.append(50*(n+m)/len(school_data))
    
    Index.append(school)
    school_type.append(school_data.loc[:, 'type'].unique()[0])
    total_students.append(len(school_data))
    budget.append(school_data.loc[:, "budget"].unique()[0])
    average_math_score.append(school_data.loc[:, "math_score"].mean())
    average_reading_score.append(school_data.loc[:, "reading_score"].mean())
    
 
    
per_student_budget = [i / j for i, j in zip(budget, total_students)] 

df=pd.DataFrame(
    {"": Index,
     "School Type": school_type,
     "Total Students": total_students,
     "Total School Budget": budget,
     "Per Student Budget": per_student_budget,
     "Average Math Score": average_math_score,
     "Average Reading Score": average_reading_score,
     "% Passing Math": passing_math, 
     "% Passing Reading": passing_reading,
     "% Overall Passing Rate": overall_passing_rate 
    })
    
```

```python
dfNew = df.sort_values(by=['% Overall Passing Rate'], ascending=False).set_index('')

dfNew["Spending Ranges (Per Student)"] = pd.cut(dfNew["Per Student Budget"], spending_bins, labels=group_names)


Index=[]
average_math_score = []
average_reading_score = []
passing_math = []
passing_reading = []
overall_passing_rate = []

for group_name in group_names:
    dfSpendingGroups=dfNew.loc[dfNew['Spending Ranges (Per Student)']==group_name]
    Index.append(group_name)
    average_math_score.append(dfSpendingGroups['Average Math Score'].mean())
    average_reading_score.append(dfSpendingGroups['Average Reading Score'].mean())
    passing_math.append((dfSpendingGroups['% Passing Math']*dfSpendingGroups['Total Students']).sum()/(dfSpendingGroups['Total Students'].sum()))
    passing_reading.append((dfSpendingGroups['% Passing Reading']*dfSpendingGroups['Total Students']).sum()/(dfSpendingGroups['Total Students'].sum()))
overall_passing_rate=[sum(x)/2 for x in zip(passing_math, passing_reading)]
df_spending=pd.DataFrame(
    {"Spending Ranges (Per Student)": Index,
     "Average Math Score": average_math_score,
     "Average Reading Score": average_reading_score,
     "% Passing Math": passing_math,
     "% Passing Reading": passing_reading,
     "% Overall Passing Rate": overall_passing_rate
    })

df_spending.set_index('Spending Ranges (Per Student)')
```
